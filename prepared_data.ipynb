```python
import pandas as pd

df = pd.read_csv(r'/Users/dimka/Desktop/проект/load_data.csv')
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
encoded_schedule = encoder.fit_transform(df["schedule"].values.reshape(-1, 1)).toarray()

schedule_names = encoder.get_feature_names_out(['schedule'])
schedule_df = pd.DataFrame(encoded_schedule, columns=schedule_names)
df = pd.concat([df, schedule_df], axis=1)
df.drop("schedule", axis=1, inplace=True)
df
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
encoded_employment = encoder.fit_transform(df["employment"].values.reshape(-1, 1)).toarray()

employment_names = encoder.get_feature_names_out(['employment'])
employment_df = pd.DataFrame(encoded_employment, columns=employment_names)
df = pd.concat([df, employment_df], axis=1)
df.drop("employment", axis=1, inplace=True)
df
```

```python
df.drop("experience", axis=1, inplace=True)
```

```python
encoder = OneHotEncoder()
encoded_area = encoder.fit_transform(df["area"].values.reshape(-1, 1)).toarray()

area_names = encoder.get_feature_names_out(['area'])
area_df = pd.DataFrame(encoded_area, columns=area_names)
df = pd.concat([df, area_df], axis=1)
df.drop("area", axis=1, inplace=True)
df
```

```python
df["schedule_Полный день"] = df['schedule_Полный день'].astype('int')
df["schedule_Удаленная работа"] = df['schedule_Удаленная работа'].astype('int')
df["employment_Полная занятость"] = df['employment_Полная занятость'].astype('int')
df["employment_Стажировка"] = df['employment_Стажировка'].astype('int')
df["area_Москва"] = df['area_Москва'].astype('int')
df.dtypes
```

```python
df.to_csv(r'/Users/dimka/Desktop/проект/prepared_data/prepared_data.csv', index = False)
```
